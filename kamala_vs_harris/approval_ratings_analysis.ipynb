```python
congress_df = app_avg_df[app_avg_df['politician/institution'] == 'Congress']
congress_df.head()
```

```python
import plotly.express as px 

fig = px.line(congress_df, x='date', y='pct_estimate', color='answer')
fig.show()

```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Get the list of unique institutions
institutions = app_avg_df['politician/institution'].unique()

# Create a 2x2 grid for subplots
fig = make_subplots(rows=2, cols=2, subplot_titles=institutions)

# Define row and column mapping
row_col_map = [(1, 1), (1, 2), (2, 1), (2, 2)]

# Loop through each institution and add a line plot
for i, institution in enumerate(institutions):
    # Filter data for the current institution
    df_filtered = app_avg_df[app_avg_df['politician/institution'] == institution]

    # Add traces for approval and disapproval
    for answer in df_filtered['answer'].unique():
        subset = df_filtered[df_filtered['answer'] == answer]

        # Set color based on the answer
        color = 'lightgreen' if answer.lower() == 'approve' else 'salmon'
        legend_name = 'Approve' if answer.lower() == 'approve' else 'Disapprove'

        # Add trace with simplified legend
        fig.add_trace(
            go.Scatter(
                x=subset['date'],
                y=subset['pct_estimate'],
                mode='lines',
                line=dict(color=color),
                showlegend=(i == 0),  # Only show legend once
                name=legend_name
            ),
            row=row_col_map[i][0],
            col=row_col_map[i][1]
        )

# Update layout with legend settings
fig.update_layout(
    title="Approval Ratings for Different Institutions",
    xaxis_title="Date",
    yaxis_title="Percentage Estimate",
    height=800,
    showlegend=True,
    legend=dict(
        itemsizing='constant',
        title="Legend"
    )
)

# Show the figure
fig.show()
```
